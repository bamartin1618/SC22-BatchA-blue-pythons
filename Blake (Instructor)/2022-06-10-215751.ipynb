```python
import pandas
from sklearn.preprocessing import LabelEncoder
```

```python
instructor_dataset = pandas.read_csv('instructor_data.csv')

student_dataset = pandas.read_csv('student_data.csv')
student_dataset.columns = [i for i in range(student_dataset.shape[1])]

student_dataset[7] = student_dataset[7].map(
    
    {"I have never coded before! (that's TOTALLY ok :) )" : 0,
     'I have some basic programming experience, I know how to add things together and print data' : 1,
     "I've been coding for a bit, I can use for loops and if-statements" : 2,
     'I am an experienced coder, I know how to write and use functions, external packages, and more' : 3,
     'From a teacher or school administrator' : 1,
     "I have some basic programming experience, I know how to add things together and print data, I've been coding for a bit, I can use for loops and if-statements" : 1,
     "I've been coding for a bit, I can use for loops and if-statements, I am an experienced coder, I know how to write and use functions, external packages, and more": 2,
     "I have some basic programming experience, I know how to add things together and print data, I've been coding for a bit, I can use for loops and if-statements, I am an experienced coder, I know how to write and use functions, external packages, and more": 1,
     "I have never coded before! (that's TOTALLY ok :) ), I have some basic programming experience, I know how to add things together and print data": 0,
     "I have never coded before! (that's TOTALLY ok :) ), I have some basic programming experience, I know how to add things together and print data, I've been coding for a bit, I can use for loops and if-statements" : 0
    }
    
)
student_dataset = student_dataset[[0, 7]]
student_dataset = student_dataset[(student_dataset[7] == 0) | (student_dataset[7] == 1) | (student_dataset[7] == 2) | (student_dataset[7] == 3)]

student_dataset[0] = student_dataset[0].map(lambda name: name.split(" ")[0].capitalize())
student_dataset.columns = ['firstname', 'skillcategory']

student_dataset
```

```python
instructor_dataset = instructor_dataset[['firstname', 'average']]
instructor_dataset
```

```python
student_dataset = student_dataset.groupby('firstname').mean()

instructor_dataset = instructor_dataset.join(student_dataset, on = 'firstname')

instructor_dataset = instructor_dataset.dropna(0, 'any')

instructor_dataset
```

```python
from scipy.stats import pearsonr

print(pearsonr(instructor_dataset['average'], instructor_dataset['skillcategory']))
```
