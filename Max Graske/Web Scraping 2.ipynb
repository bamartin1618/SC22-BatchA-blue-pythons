```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://www.w3schools.com/python/default.asp')
html = response.text

parser = BeautifulSoup(html, 'html.parser')

desired_data = parser.find('div', attrs = {'id' : 'leftmenuinnerinner'})

lesson_urls = desired_data.find_all('a')

url_first_part = 'https://www.w3schools.com/python/'

for lesson_url in lesson_urls:
    print(url_first_part + lesson_url['href'])
```
